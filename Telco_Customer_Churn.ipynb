```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "Telco-Customer-Churn.csv"
file_data = pd.read_csv(file_name, index_col = "customerID")
```

```python
file_data.head()
```

```python
file_data.info()
```

```python
file_data['SeniorCitizen'] = file_data.SeniorCitizen.astype('object')
file_data.info()
```

```python
file_data['TotalCharges'] = pd.to_numeric(file_data.TotalCharges, errors = 'coerce')
file_data.info()
```

```python
file_data.TotalCharges.fillna(value = file_data.tenure *  file_data.MonthlyCharges)
file_data.info()
```

```python
file_data.describe()
```

```python
plt.figure(figsize = (7,7))
sns.regplot(x = 'tenure', y = 'TotalCharges', data = file_data)
```

```python
#Total charges for the customer keep increasing the longer the tenure
```

```python
file_data['Total_online_services'] = (file_data[['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']] == 'Yes').sum(axis=1)
```

```python
plt.figure(figsize = (12,5))
sns.countplot(x = 'Total_online_services', hue = 'Churn', data = file_data)
```

```python
# Customers who opt for fewer online services tend to churn more as compared to those who have opted for multiple online services.
# But it could also be that most people opt for only 1 service and hence the churn is more. We will analyze this by plotting the churn percentages
```

```python
churn_count = file_data.replace('Yes',1).replace('No', 0).groupby('Total_online_services', as_index=False)[['Churn']].mean()
churn_count[['Churn']] = np.round(churn_count[['Churn']], 2)*100
```

```python
plt.figure(figsize = (12,5))
sns.barplot(x = 'Total_online_services', y = 'Churn', data = churn_count)
```

```python
#Customers who does not avail any internet services are churning less as compared to 1 service.  
#Customers who are availing just one Online Service are churning highest. 
#As the number of online services increases beyond one service, the less is the proportion of churn
```

```python
telecom_churn_services = file_data[['OnlineSecurity', 'DeviceProtection', 'StreamingMovies'
                                       ,'TechSupport', 'StreamingTV', 'OnlineBackup', 'Churn']]
telecom_churn_services.replace(to_replace='Yes', value=1, inplace=True)
telecom_churn_services.replace(to_replace='No', value=0, inplace=True)
telecom_churn_services = telecom_churn_services[telecom_churn_services.OnlineSecurity !='No internet service']             
agg = telecom_churn_services.groupby('Churn', as_index=False)[['OnlineSecurity', 'DeviceProtection', 'StreamingMovies', 'TechSupport', 'StreamingTV', 'OnlineBackup']].sum()
```

```python
ax = agg.set_index('Churn').T.plot(kind='bar', stacked=True, figsize=(12,6))
patches, labels = ax.get_legend_handles_labels()
ax.legend(patches, labels, loc='best')
ax.set_title('Which Service Customers Churn Higher', fontsize=20)
```

```python
#Customers who are availing Streaming Movies and StreamingTV are churning in higher proportions
```

```python
plt.figure(figsize=(12,6))
ax = sns.boxplot(x='Churn', y = 'tenure', data=file_data)
ax.set_title('Churn vs Tenure', fontsize=20)
ax.set_ylabel('Tenure', fontsize = 15)
ax.set_xlabel('Churn', fontsize = 15)
```

```python
#Customers with shorter tenure tend to churn out more often
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x="Churn", hue="SeniorCitizen", data=file_data);
ax.set_ylabel('Number of Customers', fontsize = 15)
ax.set_xlabel('Churn', fontsize = 15)

ax.set_title('Churn Among Senior Citizens vs Non Senior (Absolute Values)', fontsize=20)
```

```python
agg = file_data.replace('Yes', 1).replace('No', 0).groupby('SeniorCitizen', as_index=False)[['Churn']].sum()
agg.iloc[0,1] = round(agg.iloc[0,1]/file_data[file_data.SeniorCitizen==0].shape[0], 2) * 100
agg.iloc[1,1] = round(agg.iloc[1,1]/file_data[file_data.SeniorCitizen==1].shape[0], 2) * 100
plt.figure(figsize=(12,6))
ax = sns.barplot(x='SeniorCitizen', y ='Churn', data = agg)
ax.set_ylabel('Percentage of Churn', fontsize=15)
ax.set_xlabel('Senior Citizen', fontsize=15)

ax.set_title('Churn Among Senior Citizens vs Non Senior (Percentage)', fontsize=20)
```

```python
# Majority of senior citizens tend to churn
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x="SeniorCitizen", data=file_data);
ax.set_ylabel('Number of Customers', fontsize=15)
ax.set_xlabel('Senior Citizen', fontsize=15)

ax.set_title('Senior Citizen or Not', fontsize=20)
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x="Churn", hue="gender", data=file_data);
ax.set_ylabel('Number of Customers', fontsize = 15)
ax.set_xlabel('Churn', fontsize = 15)

ax.set_title('Churn By Gender', fontsize=20)
```

```python
agg = file_data.replace('Yes',1).replace('No', 0).groupby('Total_online_services', as_index=False)[['MonthlyCharges']].mean()
agg[['MonthlyCharges']] = np.round(agg[['MonthlyCharges']], 0)

plt.figure(figsize=(12,6))
ax = sns.barplot(y='MonthlyCharges', x='Total_online_services', data=agg)
ax.set_xlabel('Number of Online Services Availed', fontsize=15)
ax.set_ylabel('Average Monthly Charges',  fontsize=15)
ax.set_title('Avg Monthly Charges vs Number of Services', fontsize=20)
```

```python
#Customers who have availed multiple online services pay more than those with fewer online services
#Customers with 6 online services pay almost double compared to those with just 1 online service
```

```python
agg = file_data.replace('Yes', 1).replace('No', 0).groupby('tenure', as_index=False)[['Churn']].mean()
agg = agg[agg.tenure < 25]
agg['Churn'] = np.round(agg['Churn'], 2) * 1000

plt.figure(figsize=(20,10))

ax = sns.barplot(x='tenure', y='Churn', data = agg)
ax.set_title('Churn Percentage Over 24 Months of Tenure', fontsize=20)
ax.set_ylabel('Percentage of Churn', fontsize = 15)
ax.set_xlabel('Tenure in Months', fontsize = 15)
```

```python
# Most customers tend to churn in the first month of their tenure
# As the months pass, the number of customers that churn decreases
```
