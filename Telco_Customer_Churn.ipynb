```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_name = "Telco-Customer-Churn.csv"
file_data = pd.read_csv(file_name, index_col = "customerID")
```

```python
file_data.head()
```

```python
file_data.info()
```

```python
file_data['SeniorCitizen'] = file_data.SeniorCitizen.astype('object')
file_data.info()
```

```python
file_data['TotalCharges'] = pd.to_numeric(file_data.TotalCharges, errors = 'coerce')
file_data.info()
```

```python
file_data.TotalCharges.fillna(value = file_data.tenure *  file_data.MonthlyCharges)
file_data.info()
```

```python
file_data.describe()
```

```python
plt.figure(figsize = (7,7))
sns.regplot(x = 'tenure', y = 'TotalCharges', data = file_data)
```
